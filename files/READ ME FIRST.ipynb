A series of interactive notebooks for exploring simple geolocation related issues originally prepared in partial support of the Open University module [*Introduction to computing and information technology 2 (TM112)*](http://www.open.ac.uk/courses/modules/tm112).

This environment runs completely in the browser using `jupyterlite`. Python code execution is supported using a WASM (Web Assembly) `pyodide` powered Python environment.

*Note that the in-browser `jupyterlite` environment is under active development. Not all the functionality provided by a complete Python/IPython environment is available.*

Navigate to each of the activity notebooks in turn by double clicking on the appropriate file in the file browser in the left hand sidebar.

*For more information on how to use Jupyter notebooks, see the notebook `Using Jupyter Notebooks.ipynb`.*
