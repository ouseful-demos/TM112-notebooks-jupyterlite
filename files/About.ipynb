# About

This set of the notebooks and the associated environment provide a demonstration of providing simple Python based teaching and learning activities using Jupyter notebooks running purely inside a browser using [`jupyterlite`](https://jupyterlite.readthedocs.io/).

No external Jupyter server is required: the environment is published a simple web server in exactly the same way as a simple, static website.

Code execution runs completely in the browser. In the demonstration notebooks, Python code execution is supported using a WASM (Web Assembly) [`pyodide` powered Python environment](https://pyodide.org/en/stable/). A full stack of Python packages are already supported, with more on the way.

Whilst the demonstration does place certain requirements on the user — a recent web browser, a good web connection to load the environment into the browser for the first time, and a computer powerful enough to run the WASM environment and the Python code executed within it — *no Python related installation requirements are required by either the end user or the publisher*.

If the user continues to use the same browser, file edits will be saved and persisted into local browser storage. Files created by the user can also be saved and edited at a later date, agina using local browser storage.

This deployment model may be of significant benefit in open educational contexts:

- compute resourcing does NOT have to be provided by the publishing organisation (a simple web server will do);
- content can be distributed to the user in the form of notebooks and other files that are used to seed the initial environment;
- users do not need to install any software — all they need is a web browser and a network connection to open the environment for the first time;
- users can create, edit and save their own files using bowser local storage;
- users can install and import additional Python packages, insofar as they are available and can run in the `pyodide` environment.

The notebooks were originally prepared in partial support of the Open University module [*Introduction to computing and information technology 2 (TM112)*](http://www.open.ac.uk/courses/modules/tm112) and describe a range of simple interactive activities for exploring simple geolocation related issues.

## See Also...

The original notebooks have also been rendered into an interactive online textbook style format using Jupyter Book. You can try it out here: [*TM112 notebooks Jupyter Book*](https://ouseful-course-containers.github.io/ou-tm112-notebooks/).

Interactivity in the form of inline code execution within the HTML site and the rendering of executed code output is provided using a MyBinder kernel and Thebelab.js.
